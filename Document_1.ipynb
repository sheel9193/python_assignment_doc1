```python
result_div = num2/num1
print(result_div)

#short-hand version
result_div /= 2
print(result_div)
```

## Using 'and'/'or' with If Statements

### Lists

##### Declaring Lists

```python
my_list = [] # Empty List Declaration

names = ['Max','Cindy','Kathy', "bob","Nate"]

print(names)
```

##### Indexing a List

```python
# [start: stop: step] -- [-1]

#Single Index
print(names[0])

# Print starting at index 1 going to the end
print(names[1:])

#Print starting at beginning of list to N (n == the number we want to stop at)
print(names[:2])

# print starting at index 1 going up BY 2 in each iteration
print(names[1::2])

# print starting at the back and present in reverse order
print(names[::-1])

# Print the last value NO MATTER THE SIZE of the list!
print(names[-1])
```

##### .append()

```python
# does not change original data, just adds to the end 
names.append('Joel')
print(names)
```

##### .insert()

```python
# list.insert(index,value)

names.insert(3,"Gabby")
print(names)
```

##### .pop()

```python
# default is the last value if no parameter is given
# pop returns the element that was removed in case you want to assign it into a variable

# my_name = names.pop()
# print(my_name)

my_name = names.pop(3)
print(my_name)
print(names)
```


##### .remove()

```python
# value to be removed , not index
# names.remove("Bob")
# print(names)

# remove multiple bobs out of list
while 'bob' in names:
    names.remove('bob')
print(names)
```

##### del()

```python
# index to be removed, not value
# BE CAREFUL with del, can cause indexing errors if not used carefully
del(names[1])
print(names)
```

##### Lists Within Lists

```python
# Lists are able to hold any type of element, including OTHER lists themsevles
# they can also go as deep as you want, having many lists nested inside of each other

names = ["Max", "Sam","Josh",["Sally","Sue","Tameka"]]

print(names)
print(names[3][1])
```

##### Looping Through Lists

```python
# two ways to loop through a list, one is by index, the other is by using the 'in' keyword

# With a for loop using the index 
for i in range(len(names)):
    print(names[i])
    
# Loop with a for loop (Using the 'in' keyword)
for i in names:
    print(i)
```

## Exercise #1 <br>
<p>Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.</p>

```python
list_num = []

number = 1


while True:
    number_cubed = number ** 3
    if number_cubed <= 1000:
        list_num.append(number_cubed)
        number += 1
    else:
        break

print (list_num)




        
        
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

def prime():
    prime = []

    for number in range (2,101):
        for index in range (2,number):
            if number % index == 0:
                break
        else:
            prime.append(number)
    return prime
prime()

```
